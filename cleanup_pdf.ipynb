```python
import pandas as pd
```

```python
sicknesses = pd.read_csv("Sicknesses.csv")
```

```python
sicknesses = sicknesses.drop(columns=['Prevention & Control Measures'])
print(sicknesses.head())
```

```python
# verificar rows q tenham o primeiro elemento "Disease/ Illness" e elimina-los
sicknesses = sicknesses[sicknesses.iloc[:, 0] != 'Disease/ Illness']
print(sicknesses)
```

```python
sicknesses = sicknesses.drop(columns=['Diseases'])
print(sicknesses.head())
```

```python
sicknesses_clean = sicknesses.dropna()

sicknesses_clean.to_csv('Sicknesses_clean.csv', index=False)
```
